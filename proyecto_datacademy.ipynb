Habiendo transcurrido dos años de pandemia, el mercado retail en los Estados Unidos se encuentra en una etapa de crecimiento, ya que en 2021 [registró el doble de aperturas de tiendas frente a las que se cerraron](https://nrf.com/blog/us-retailers-announced-many-store-openings-in-2021). Esto ofrece un ambiente propicio para proyectar a futuro la creación de nuevas marcas y empresas. Además, este sector es atractivo para el mercado bursátil que generalmente suele acaparar entre el [10 y el 14% del marketshare,](https://www.eleconomista.es/mercados-cotizaciones/noticias/10922415/12/20/El-inversor-retail-de-EEUU-se-apodera-ya-del-20-del-mercado-en-pleno-apogeo-de-la-pandemia.html) lo que podría ofrecer buenas fuentes de captación de capitales. Teniendo esto en cuenta se realizó un análisis de los 25 mayores retails de Estados Unidos con el fin de evaluar la posibilidad de ingreso en el secto

El mayor actor en el mercado es, sin dudas, Walmart, quien domina con el 37% de las ventas totales, diferenciándose de la empresa con menos ingresos en US\$641,527.00 millones. Walmart pertenece a la categoría de "Supercenters", por lo que intentar ingresar en este nicho implica tener como competencia a un gigante. Si bien no sería imposible prosperar allí, se continuó analizando en busca de otras alternativas.

Para el resto de los competidores, quitando a Walmart del análisis, se observó una mediana de ventas que ronda los US\$ 34.694,0 millones, con un mínimo de US\$ 16.592 y un máximo de US\$ 115.037 millones \(\$98,445.00 millones de diferencia\). Se pudo apreciar una gran concentración de las ventas entre el mínimo y la mediana. La mayoría de las empresas tuvo ingresos entre US\$ 21,926 y 74,129 millones, sumando en total US\$ 575,980 millones \(Gráficos 2 y 3\).


Con respecto a la infraestructura inmobiliaria la mayoría de las empresas posee entre 623 y 4.557 tiendas \(Gráficos 4 y 5\), a excepción de 3 compañías con más de 9.000 tiendas (Dollar Tree con 14.250, Dollar General con 13.350 y CVS con 9.813) y dos compañías que operan exclusivamente online \(Amazon y Apple\). Un punto importante a destacar acerca del número de tiendas, es que la cantidad de las mismas no implica necesariamente un mayor número de ingresos \(Gráfico 6\). Tal es así que las empresas que operan online superan a 2 de las que más tiendas tienen \(pertenecientes a la categoría Dollar Stores\) y se encuentran por encima de la mediana en cuanto a ventas \(Gráfico 10\). Este es un aspecto a evaluar si lo que se busca es invertir en nuevas tecnologías y canales de venta alternativos.

Walgreen Boots y Walmart US completan las 5 compañías con más locales (junto con las 3 empresas con más de 9.000 tiendas mencionadas en el parráfo anterior) y son a su vez parte de los 5 mayores vendedores \(Gráfico 7\).

Entre las empresas que sacan mejor partido de cada tienda, se encuentra liderando Cosco con US\$187.80 millones por tienda promedio \(Gráfico 9\).

Se analizaron las categorías a fin de evaluar el nicho más conveniente para desarrollar el nuevo negocio. Se encontró que a excepción de los "Supercenters" \(US\$745.919 millones en ventas\), la categoría que más ingresos generó fue "Grocery" \(US\$297.164 millones\) seguida por "Drug Stores", "Warehouse Club" y "Home Improvement" \(Gráfico 8\). Al realizar una comparación con las categorías que menos número de tiendas tiene, se apreció que "Home Improvement" y "Warehouse  Club" se encuentran en ambos grupos \(Gráfico 11\), resultando esta última la categoría con mejores ingresos por tienda \(Gráfico 12\).

Por último, se puede decir que el sector se encuentra aumentando en promedio su presencia en tiendas físicas, principalmente los "Dollar Stores" (\Gráfico 13\) con un 10,6%. Como excepción, "Electronics" contrajo el número de tiendas en un 3,9%. 

### Conclusiones
- El mercado se encuentra fuertemente dominado por Superstores, principalmente Walmart.
- Mayor número de tiendas no implica mayores ventas.
- Las empresas con ventas online presentan ingresos por encima del promedio de las empresas que poseen tiendas físcas, aunque no es una posición preponderante.
- La categoría con más ventas por tienda promedio es "Warehouse Club" y particularmente Cosco es la empresa que destaca en ingresos por venta en el mercado.
- El sector se encuentra en crecimiento, liderando el aumento de tiendas las "Drug Stores" y "Dollar Stores".

## Diferencia en promedio de ventas

Se evalúa el error porcentual entre la división de las ventas sobre las tiendas y la columna "Sales/Avg. Store"

```python
sales_avg_company=pd.DataFrame(df_lr['Company'])
sales_avg_company['Sales/Avg. Store Diff.']= abs((1 - (df_lr['Sales']/df_lr['Stores'])/df_lr['Sales/Avg. Store']))
sales_avg_company=sales_avg_company.sort_values('Sales/Avg. Store Diff.',ascending=0).head()
sales_avg_company.style.format(formatter={'Sales/Avg. Store Diff.':"{:.2%}"})
```

```python
df_lr_promedio_ventas = df_lr.copy()
df_lr_promedio_ventas["Stores"].dropna()
df_lr_promedio_ventas['Avg. Sales'] = (df_lr['Sales/Avg. Store']*df_lr['Stores']/df_lr['Sales']*(1-df_lr["Store Count Growth"]/2))
df_lr_promedio_ventas=df_lr_promedio_ventas.dropna()
df_lr_promedio_ventas=df_lr_promedio_ventas.sort_values("Avg. Sales")
fig,ax = plt.subplots(figsize=(15,8))
ax.set_title("Gráfico 1. Correlación entre las ventas y las ventas promedio por compañía",size=20)
leyenda = """
Ventas promedio por tienda
-------------------------------------
(ventas por tienda) [totales]
"""
ax=sns.scatterplot(x=df_lr_promedio_ventas["Company"],y=df_lr_promedio_ventas["Avg. Sales"],label=leyenda)

#Eje x
ax.xaxis.set_tick_params(rotation=90,labelsize=13)
ax.set_xlabel("Compañía", size=13)

#Eje y
ax.axhline(1,0,1,linestyle="--",color="red")
ax.set_ylabel("('Sales'/'Stores')/('Sales/Avg. Stores') * c", size=13)
ax.set_ylim([0.4,1.05])
ax.yaxis.set_tick_params(labelsize=13)

#Leyenda
ax.legend(loc="lower center",fontsize=15)
plt.show()
```

_________________
- La compañía Walmart US presenta una diferencia mayor al 100% entre el promedio de ventas en la base de datos (Sales/Avg. Store) y el cociente entre el total de ventas y el número de tiendas (Sales/Stores), por lo que se decide utilizar este último valor.
- En situaciones donde el "Store Count Growth" es alto, la relación entre "Sales/Avg. Store" y "Sales/Stores" tendrá un valor alejado de 1. Lo que explicaría, por ejemplo el caso de "CVS incl.Target" que tuvo un crecimiento del 25,6%. Esta situación no explica el desvío de Walmart (1,3% de crecimiento en tiendas).
- El coeficiente "c" en el eje 'y' corresponde a (1-"Store Count Grouth"/2).
